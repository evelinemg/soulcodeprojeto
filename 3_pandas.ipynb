## Enunciado do Problema

Analise do Comportamento do Consumidor é uma análise detalhada do consumidor ideal de uma empresa. Ajuda o negócio a compreender seus consumidores e faz com que seja fácil para eles modificar os produtos de acordo com as necessidade especificas, comportamentos e preocupações de diferentes tipos de consumidores.

Analise Comportamento do consumidor ajuda um negócio a modificar seus produtos baseado no público-alvo de diferentes segmentos de consumidores. Por exemplo, no lugar de investir dinheiro para divulgar um novo produto para todos os consumidores que estão no bancos de dados da empresa, pode-se fazer uma análise de qual segmento de consumidores provavelmente compraria o produto e então divulgar o produto somente para esse segmento específico.
